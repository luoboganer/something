# 1. 样式与分布图

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## seaborn样式
* 默认的样式有 dark darkgrid white whitegrid ticks 五种
* sns.set()方法可以设置多种参数

```python
# 默认样式
years=list(range(1950,2020,10))
gdp=[300.2,543.3,1075.0,2862.5,5979.6,10287.9,14958.3]
plt.scatter(years,gdp,color='red',marker="o")
plt.show()
```

```python
# 设置darkgrid样式
sns.set_style("darkgrid")
plt.scatter(years,gdp,color="green",marker="x")
plt.show()
```

```python
sns.set_style("dark")
plt.scatter(years,gdp,color="green",marker="D")
plt.show()
```

```python
sns.set(style="ticks",palette="muted",color_codes=True)
# stylew为样式风格，palette为调色板，color_codes为眼色代码
plt.plot(np.arange(10))
plt.show()
```

## 移除坐标轴

```python
sns.set(style="white",palette="muted",color_codes=True)
plt.plot(np.arange(10))
sns.despine()
plt.show()
```

```python
sns.set(style="white",palette="muted",color_codes=True)
plt.plot(np.arange(10))
sns.despine(offset=10,trim=True)
plt.show()
```

```python
sns.set(style="whitegrid",palette="muted",color_codes=True)
plt.plot(np.arange(10))
sns.despine(left=True,bottom=True)
plt.show()
```

## 单变量分布图

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
# 简单的单变量分布直方图与嵌套的密度函数
sns.distplot(tips["total_bill"])
plt.show()
```

```python
# 单变量分布中设置bins，去掉概率密度图kde，改用毛毯图rug
sns.distplot(tips["total_bill"],bins=30,kde=False,rug=True)
plt.show()
```

```python
# 去掉直方图而保留概率密度图和毛毯图
sns.distplot(tips["total_bill"],hist=False,rug=True,kde=True)
plt.show()
```

```python
# 绘制多个子图
fig,axes=plt.subplots(1,2,figsize=(14,6))
sns.distplot(tips["total_bill"],ax=axes[0],kde=False)
sns.rugplot(tips["total_bill"],ax=axes[0])
sns.kdeplot(tips["total_bill"],ax=axes[1],shade=True)
sns.rugplot(tips["total_bill"],ax=axes[1])
plt.show()
```

## 多变量分布图

```python
sns.jointplot(x="total_bill",y="tip",data=tips)
plt.show()
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde")
# 用等高线图代替散点图，用概率密度图代替数据分布直方图
# kind参数的选项有 scatter,reg,resid,kde,hex
plt.show()
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex")
# 用等高线图代替散点图，用概率密度图代替数据分布直方图
# kind参数的选项有 scatter,reg,resid,kde,hex
plt.show()
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg")
# 用等高线图代替散点图，用概率密度图代替数据分布直方图
# kind参数的选项有 scatter,reg,resid,kde,hex
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

# 3. 回归图与网格

## 回归图

```python
sns.jointplot(x="total_bill",y="tip",data=tips)
plt.show()
```

```python
#lmplot与regplot画出的回归图基本相同
sns.set(style="darkgrid",color_codes=True)
sns.regplot(x="total_bill",y="tip",data=tips)
plt.show()
```

```python
#使用ci参数为False取消置信区间
sns.set(style="darkgrid",color_codes=True)
sns.lmplot(x="total_bill",y="tip",data=tips,ci=False)
plt.show()
```

```python
# 使用hue参数添加分类变量
sns.set(style="darkgrid",color_codes=True)
sns.lmplot(x="total_bill",y="tip",data=tips,hue="smoker")
plt.show()
```

```python
# 再添加一个变量可以通过子图呈现
sns.lmplot(x="total_bill",y="tip",hue="smoker",col="time",data=tips)
plt.show()
```

```python
# 总共为5个变量的时候
sns.lmplot("total_bill",y="tip",hue="smoker",col="time",row="sex",data=tips)
plt.show()
```

## 网格
* 网格技术：在对多维度数据进行可视化时，在数据集的不同子集上绘制同一个绘图的多个实例

```python
g=sns.FacetGrid(tips,col="time")
g.map(plt.hist,"tip")
plt.show()
```

```python
g=sns.FacetGrid(tips,col="day",hue="smoker")
g.map(plt.scatter,"total_bill","tip",alpha=0.7)
g.add_legend()
plt.show()
```

```python
g=sns.FacetGrid(tips,col="day",row="sex")
g.map(sns.boxplot,"size","tip",order=None)
g.add_legend()
plt.show()
```

# 4. 综合示例——泰坦尼克号生还着数据

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic.head()
```

## 数据清洗

```python
titanic.isnull().sum()
```

```python
# 查看乘客的年龄分布情况
sns.set(style="darkgrid",palette="muted",color_codes=True)
sns.distplot(titanic[titanic["age"].notnull()]["age"])
plt.show()
```

```python
# 从上图可以看出年龄呈现正态分布，因此采用均值填充缺失值
titanic["age"]=titanic["age"].fillna(titanic["age"].mean())
sns.distplot(titanic["age"])
plt.show()
```

```python
# 统计显示上船地点的分布
sns.countplot(x="embarked",data=titanic)
```

```python
# 从上图可以看到大部分人在地点“S”上船，因此用“S”填充缺失值
titanic["embarked"]=titanic["embarked"].fillna("S")
# deck字段缺失值太多，将其删除,其他一下不用的字段也一起删除
titanic.drop(["survived","pclass","sibsp","parch","who","adult_male","deck","embark_town"],axis=1,inplace=True)
```

```python
titanic.head()
```

```python
titanic.isnull().sum()
```

## 数据探索

```python
fig,axes=plt.subplots(1,2,figsize=(10,6))
sns.countplot(x="sex",data=titanic,ax=axes[0])
sns.countplot(x="sex",data=titanic,hue="alive",ax=axes[1])
plt.show()
# 可以看到男乘客明显多一些,但是女乘客幸存的比例明显更大
```

```python
sns.boxplot(x="sex",y="age",data=titanic)
plt.show()
# 可以看到男女乘客年龄基本接近，但是女乘客年龄分布跨度更大
```

```python
sns.countplot(x="alive",hue="sex",data=titanic)
plt.show()
# 明显可以看到女士优先逃生
```

```python
# 利用网格技术进行更明显的对比
g=sns.FacetGrid(titanic,col="sex")
g.map(sns.countplot,"alive")
plt.show()
```

```python
# 老人和孩子也是优先考虑的对象，因此对年龄进行分级，观察他们的额情况
def ageLevel(age):
    if age<=16:
        return "child"
    elif age >=60:
        return "aged"
    else:
        return "midlife"
titanic["ageLevel"]=titanic["age"].map(ageLevel)
```

```python
# 年龄分级后的数据效果
titanic.head()
```

```python
sns.countplot(x="ageLevel",data=titanic)
plt.show()
# 分级后的年龄分布，可以看到成年人还是占绝了乘客的绝大部分
```

```python
# 乘客年龄分布与生还与否的关系:基本没有联系，但是小孩生还率略高
sns.countplot(x="alive",hue="ageLevel",data=titanic)
plt.show()
```

```python
df=titanic.groupby(["ageLevel","alive"])["alive"].agg("count")
```

```python
count=df.unstack()
```

```python
count
```

```python
count["aliveRate"]=count["yes"]/(count["no"]+count["yes"])
```
