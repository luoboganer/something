## pd.read_csv()与pd.read_table()的参数说明
* path 文件路径
* sep 分隔符，即字段隔开的字符序列，也可以使用正则表达式
* header 指定列索引，默认为行索引0的行（第一行），也可以为None，也可以通过names参数手动指定
* index_col 用于行索引的列名或者列编号
* names 手动指定列索引的列名
* skiprows 需要忽略的行数（从文件开始处算起）
* nrows 需要读取的行数（从文件开始处算起）
* chunksize 文件分块读入时分块的大小（单位是行）
* useclos 指定需要读取的列

## 网络爬虫

```python
#!/usr/bin/env python 
# -*- coding:utf-8 -*- 
'''
 * @Author: shifaqiang(石发强)--[14061115@buaa.edu.cn] 
 * @Date: 2018-10-29 11:09:14 
 * @Last Modified by:   shifaqiang 
 * @Last Modified time: 2018-10-29 11:09:14 
 * @Desc: a simple spider for http://www.kugou.com/yy/rank/home/1-8888.html?from=rank
'''

import requests
import bs4

data=[]
url="http://www.kugou.com/yy/rank/home/1-22163.html?from=rank"
web_data=requests.get(url)
soup=bs4.BeautifulSoup(web_data.text,"lxml")
ranks=soup.select("span.pc_temp_num")
titles=soup.select("div.pc_temp_songlist > ul > li > a")
times=soup.select("span.pc_temp_tips_r > span")

for rank,title,time in zip(ranks,titles,times):
    a={
        "rank":rank.get_text().strip(),
        "singer":title.get_text().split("-")[0],
        "song":title.get_text().split("-")[1],
        "time":time.get_text().strip(),
    }
    data.append(a)
```

```python
data
```

```python
df23=pd.DataFrame(data)
```
