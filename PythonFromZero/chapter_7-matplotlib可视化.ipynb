## 线条颜色与线型
* color参数指定线条颜色，可以用RGB值表示，也可以用英文单词描述
* linestyle参数指定线型

```python
years=[1950,1960,1970,1980,1990,2000,2010]
gdp=[300.2,545.3,1075.9,2862.5,5979.6,10289.7,14958.3]
plt.plot(years,gdp,color="red")
```

```python
plt.plot(years,gdp,linestyle="--",color="green")
```

```python
plt.plot(years,gdp,linestyle="--",marker="x",markersize=10,markeredgecolor="red",color="blue",linewidth=5)
```

# 2. 柱状图

```python
data=[23,85,72,43,52]
plt.bar([1,2,3,4,5],data)
```

```python
plt.bar(range(len(data)),data,color="royalblue",alpha=0.7)
```

```python
plt.bar(range(len(data)),data,color="green",alpha=0.7)
plt.grid(color="#95a5a6",linestyle=":",linewidth=1,axis="y",alpha=0.5)
```

```python
# 堆积柱状图
data1=[23,85,72,43,52]
data2=[42,35,21,16,9]
plt.bar(range(len(data1)),data1)
plt.bar(range(len(data2)),data2,bottom=data1)
```

```python
# 并列柱状图
data1=[23,85,72,43,52]
data2=[42,35,21,16,9]
width=0.3
plt.bar(np.arange(len(data1)),data1,width=width)
plt.bar(np.arange(len(data2))+width,data2,width=width)
```

```python
# 水平柱状图
data1=[23,85,72,43,52]
plt.barh(range(len(data1)),data1)
```

```python
data=[23,85,72,43,52]
labels=list("ABCDE")
# 设置刻度
plt.xticks(range(len(data)),labels)
# 设置轴标签
plt.xlabel("Class")
plt.ylabel("Amounts")
plt.title("Bar Example")
plt.bar(range(len(data)),data)
plt.show()
```

```python
data1=[23,85,72,43,52]
data2=[42,35,21,16,9]
width=0.3
plt.bar(np.arange(len(data1)),data1,label="priceOne",width=width)
plt.bar(np.arange(len(data2))+width,data2,label="priceTwo",width=width)
plt.legend()
plt.xticks(range(len(data1)),list("ABCDE"))
plt.xlabel("class")
plt.ylabel("price")
plt.title("Bar Example 2")
plt.show()
```

## 文本注解

```python
data=[23,85,72,43,52]
labels=list("ABCDE")
plt.xticks(range(len(data)),labels)
plt.xlabel("Class")
plt.ylabel("Amounts")
plt.title("Example")
plt.bar(range(len(data)),data,color="r")
for x,y in zip(range(len(data)),data):
    plt.text(x,y,y,ha="center",va="bottom",color="green")
    # 根据坐标x,y确定文字位置
plt.show()
```

## 样式与字体

```python
## 使用类R语言的“ggplot”样式
data=[23,85,72,43,52]
labels=list("ABCDE")
plt.style.use("ggplot")
plt.xticks(range(len(data)),labels)
plt.xlabel("Class")
plt.ylabel("Amounts")
plt.title("Example")
plt.bar(range(len(data)),data,color="r")
for x,y in zip(range(len(data)),data):
    plt.text(x,y,y,ha="center",va="bottom",color="green")
    # 根据坐标x,y确定文字位置
plt.show()
```

```python
# 中文乱码示例
plt.title("案例")
plt.show()
```

### starbucks在不同国家的分布情况

```python
country_count=starbucks["Country"].value_counts()[:10]
```

```python
country_count
```

```python
# 使用pandas内置的绘图
country_count.plot(kind="bar",figsize=(10,6),title="星巴克分布国家Top10柱状图")
plt.show()
```

```python
# 使用matplotlib手动绘图
fig=plt.figure(figsize=(10,6))
plt.style.use("default")
plt.xlabel("Country")
plt.ylabel("Count")
plt.title("Country Top 10")
labels=list(country_count.index)
plt.bar(range(len(labels)),country_count)
plt.xticks(range(len(labels)),labels,fontsize=12)
for x,y in zip(range(len(labels)),country_count.values):
    plt.text(x,y,y,color="green",ha="center",va="bottom")
plt.show()
```

### Starbucks在不同城市的分布情况

```python
city_count=starbucks["City"].value_counts()[:10]
```

```python
city_count
```

```python
plt.figure(figsize=(10,6))
plt.rcParams["font.sans-serif"]=["simhei"]
plt.rcParams["axes.unicode_minus"]=False
labels=list(city_count.index)
plt.xlabel("city")
plt.ylabel("count")
plt.title("City Top 10")
plt.bar(range(len(labels)),city_count,color="red")
plt.xticks(range(len(labels)),labels)
for x,y in zip(range(len(labels)),city_count.values):
    plt.text(x,y,y,ha="center",va="bottom",fontsize=15,color="green")
plt.show()
```

## starbucks在中国的分布情况

```python
china_data.head()
```

```python
china_data.shape
```

```python
china_city_count=china_data["City"].value_counts()[:10]
```

```python
china_city_count
```

```python
plt.rcParams["font.sans-serif"]=["simhei"]
plt.rcParams["axes.unicode_minus"]=False
labels=list(china_city_count.index)
plt.xlabel("City")
plt.ylabel("Count")
plt.title("星巴克在中国各城市的分布")
plt.barh(range(len(labels)),china_city_count)
plt.yticks(range(len(labels)),labels)
for x,y in zip(china_city_count.values,range(len(labels))):
    plt.text(x,y,x,ha="left",va="center",color="red")
plt.show()
```
