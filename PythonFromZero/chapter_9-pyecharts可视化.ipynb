## 折线图

```python
years=list(range(1950,2020,10))
gdp=[300.2,543.3,1075.9,2862.5,5979.6,10289.7,14958.3]
assert len(years)==len(gdp)
line=pyecharts.Line("折线统计图")
line.add("GDP",x_axis=years,y_axis=gdp,mark_ponit=["average"])
```

# 2. 其他图表

## 饼图
* 用于表现不同类别的占比情况

```python
data=[45,76,35,47,56]
labels=["电脑","手机","冰箱","彩电","洗衣机"]
pie=pyecharts.Pie("饼图")
pie.add("example",labels,data,is_label_show=True)
```

```python
data=[45,76,35,47,56]
labels=["电脑","手机","冰箱","彩电","洗衣机"]
pie=pyecharts.Pie("圆环图")
pie.add("example",labels,data,radius=[45,70],is_label_show=True)
```

```python
# 通过设置rosetype参数为area和radius两种类型绘制两种玫瑰图
data=[45,76,35,47,56]
labels=["电脑","手机","冰箱","彩电","洗衣机"]
pie=pyecharts.Pie("玫瑰图")
pie.add("radius",labels,data,radius=[20,75],center=[25,50],rosetype="radius")
pie.add("area",labels,data,radius=[20,75],center=[75,50],rosetype="area",is_label_show=True)
```

# 3. 综合示例——嗅事百科用户数据

```python
import numpy as np
import pandas as pd
import pyecharts
```

```python
data3=pd.read_csv("./data/qiushi_info.csv",encoding="utf-8")
```

```python
data3.head()
```

```python
data4=pd.read_csv("./data/user_info.csv",encoding="utf-8")
```

```python
data4.head()
```

## 数据清洗

```python
data3.info()
```

```python
data3["age"].unique()
```

```python
# 看到年龄中有“不详”的情况，并是object类型，因此先用0替换不详后，转为int类型，然后用平均值填充0
data3["age"].replace("不详",np.NAN,inplace=True)
data3["age"]=data3["age"].astype("float")
data3["age"]=data3["age"].fillna(data3["age"].mean())
data3["age"]=data3["age"].astype("int32")
```

```python
data3.head()
```

```python
data3.age.unique()
```

```python
data3.dtypes
```

```python
data4.isnull().sum()
```

```python
data4.shape
```

```python
data4.dropna(inplace=True)
```

```python
data4.info()
```

```python
data4.head()
```

```python
data4.fans=data4.fans.astype("int32")
data4.topic=data4.topic.astype("int32")
data4.qiushi=data4.qiushi.astype("int32")
data4.comment_1=data4.comment_1.astype("int32")
data4.favour=data4.favour.astype("int32")
data4.handpick=data4.handpick.astype("int32")
data4["provice"]=data4["home"].str.split(" · ").str[0]
data4.qiushi_age=data4.qiushi_age.str.strip('天').astype("int64")
```

```python
data4.head()
```

## 数据探索

```python
# 找到好笑laugh和评论数较多的数据
laugh_sort=data3.sort_values(by=["laugh"],ascending=False)[:10]
```

```python
laugh_sort
```

```python
# 绘制好笑数前十条的柱状图
bar=pyecharts.Bar("搞笑段子用户排名")
bar.add("好笑数",list(laugh_sort["id"]),laugh_sort["laugh"],
        is_label_show=True,mark_point=["max","min"],mark_line=["average"],
        is_xasixlabel_align=True,xaxis_interval=0,xaxis_name_size=12,xaxis_rotate=25)
# is_xasixlabel_align=True 设置x轴标签与与x轴刻度对应
# xaxis_interval=0 没有间隔，即每个x轴标签都显示
# xaxis_name_size=12 设置x轴标签大小
# xaxis_rorate=30 设置x轴标签旋转角度
```

```python
comment_sort=data3.sort_values(by=["comment"],ascending=False)[:10]
# 绘制好笑数前十条的柱状图
bar=pyecharts.Bar("搞笑段子用户排名")
bar.add("评论数",list(comment_sort["id"]),comment_sort["comment"],
        is_label_show=True,mark_point=["max","min"],mark_line=["average"],
        is_xasixlabel_align=True,xaxis_interval=0,xaxis_name_size=12,xaxis_rotate=25)
# is_xasixlabel_align=True 设置x轴标签与与x轴刻度对应
# xaxis_interval=0 没有间隔，即每个x轴标签都显示
# xaxis_name_size=12 设置x轴标签大小
# xaxis_rorate=30 设置x轴标签旋转角度
```

```python
data3.head()
```

```python
# 删除data3中缺失user_url的数据(匿名用户)
data3.dropna(inplace=True)
# 合并两张表
data5=pd.merge(data3,data4,on="user_url")
# 在合并后的数据中删除id重复数据
data5.drop_duplicates(["id"],inplace=True)
```
